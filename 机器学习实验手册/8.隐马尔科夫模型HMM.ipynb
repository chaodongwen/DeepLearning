```python
import jieba
import jieba.posseg
from snownlp import SnowNLP
```

```python
#使用jieba

#打开文件，并进行分词
filein = open("novel.txt", "r", encoding='utf-8')
str =filein.read()
seg = jieba.posseg.cut(str)
l = []
for i in seg:
    l.append((i.word, i.flag))
filein.close()
print(l)

#将分词结果写入文件
fout = open("novel_out.txt", "w", encoding='utf-8')
seg = jieba.posseg.cut(str)
for i in seg:
    fout.writelines([i.word, '             ', i.flag, '\n'])
fout.close()
```

```python
#使用SnowNLP

#打开文件，并进行分词
filein = open("novel.txt", "r", encoding='utf-8')
str =filein.read()
s = SnowNLP(str)
for i in s.tags:
    l.append((i[0], i[1]))
filein.close()
print(l)

#将分词结果写入文件
fout = open("novel_out_SnowNLP.txt", "w", encoding='utf-8')
s = SnowNLP(str)
for i in s.tags:
    fout.writelines([i[0], '            ', i[1], '\n'])
fout.close()
```
