```python
#1.numpy数组与python列表效率对比
import numpy as np 
import numpy.random as rand
arr = np.arange(1e7)
larr = arr.tolist()
def list_times(alist, scalar):
    for i,val in enumerate(alist):
        alist[i]=val*scalar
    return alist
#timeit 计算运行时间 timeitarr*1.1
list_times(larr,1.1)
```

```python
#5.Numpy数组的布尔操作
#Plot A
img1 = np.zeros((20,20))+3
img1[4:-4, 4:-4]=6
img1[7:-7, 7:-7]=9
print("Plot A")
print(img1)
index1 = img1>3
index2 = img1<7
compound_index =index1&index2
#compound_index =(img1>3)&(img1<7)
#Plot B
img2 = np.copy(img1)
img2[compound_index]=0
print("Plot B")
print(img2)
#Plot C
index3 = img1 == 9
index4 = (img1 == 3)|index3
img3 = np.copy(img1)
img3[index4]=0
print("Plot C")
print(img3)
#例子
#产生一个随机分布的正态函数
a = rand.randn(100) 
print(a)
index = a > 0.2 
b = a[index] 
print(b)
def suqare(num):
    return num*num
b = list(map(suqare,b))
print(b)
```

```python
#7.数学运算
#使用 numpy.dot 和 numpy.transpose 分别来进行矩阵乘法运算和矩阵转置
A = np.matrix([[3,6,-5], [1, -3, 2], [5, -1, 4]])
B = np.matrix([[12],[-2],[10]])
print(A)
print(B)
X = A*(-1)*B
print(X)
#使用数组解方程组
print("使用数组解方程组")
a = np.array([[3,6,-5],[1,-3,2],[5,-1,4]])
b = np.array([12,-2,10])
print(a)
print(b)
x = np.dot(a,np.transpose(b))
print(x)
```
